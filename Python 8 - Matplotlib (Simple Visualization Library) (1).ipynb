# MATPLOTLIB

- Matplotlib is a plotting library for python programming language and its numeric and mathamatics extension NumPy, it provides and object-oriented API for embedding plots into applications using general purpose GUI like Tkinter,wsPython,Qt, or GTK+.

## Some of the major advantages of Matplotlib are:
- generally easy to get started with simple plots
- support for custom labels and texts
- Great control of every element in a figure
- High Quality output in many formats
- Very Customizable in general

# Creating Sub Plots

```python
# Creating sub plots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'bo')
plt.subplot(2,2,3)
plt.plot(x,y,'c*-')
plt.subplot(2,2,4)
plt.plot(x,y,'g-')
```

```python
# here basically plt.subplots 2 represents number of rows, 2 represents number of columns and 1 represnt the position
# the graph is meant to be positioned and the positions can be changed
# as we have given 2 rows and 2 columns we can only get 4 graphs, similarly we can do more graphs
```

```python
# Creating a complex operation
x = np.arange(0,20)
y = x * 20 + 6
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Matplotlib Example graph')
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,4 * np.pi,0.1) # 0.1 heres is step size
y = np.sin(x)
plt.title('Creating a Sine chart')

#plot the points using matplot lib
plt.plot(x,y,'r--')
plt.show()
```

```python
# compute x and y on the datapoints for sine and consine data points
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
# subplot
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('Sin Curve')

#####
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g')
plt.title('Cosain Curve')

#Showing the figure
plt.show()
```

```python
# Creating Bar plots

x = [2,8,9]
y = [2,4,10]

x2 = [3,7,9]
y2 = [3,8,11]

plt.bar(x,y,align='center')
plt.bar(x2,y2,color = 'orange')
plt.title('Bar Graph Plotting')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

# BOX PLOT USING MATPLOTLIB

```python
data = [np.random.normal(0,std,100) for std in range (0,4)]
```

```python
# rectangular box plot 
plt.boxplot(data,patch_artist=True,vert = True);
```

# Pie CHART

```python
# Data to plot
labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = [0.1,0,0,0]

#plotting
plt.pie(sizes,labels=labels,colors=colors,explode=explode,
autopct='%1.1f%%',shadow = True)

plt.axis('equal')
plt.show()
```
