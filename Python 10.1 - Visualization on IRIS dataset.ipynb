# Practising problem on IRIS Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
# When we are comparing 1 feature we call it univariate analysis
# When there are 2 features, we call it bi variate analysis, this will creat 2 D analysis.
# This type of analysis can be done easily by seaborn
# what if we have multiple feature?, we use something called as joinplot,distplot and pairplot
# If we have more than 2 feature we need to create a 3 D diagram, but in realworld we use a pair plot.
```

```python
# correlation with Heatmap
```

```python
df.head()
# Tip is y variable
# all other features are x variable
```

```python
heat_map_corr = df.corr()
print(heat_map_corr)
```

```python
df.dtypes
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(heat_map_corr,annot=True)
```

# CORRELATION

- A Co - Realation heatmap uses colored cells, typically in a mono chromatic scale, to show a 2D correlation matrix (table) between two discrete dimentions or event types. It is very important in feature selection

# Join Plot

- A Join plot allows to study the relationship between 2 numerical variables. The central chart displays their co-relation, it is usually a scatter plot, hexbin plot, a 2D Histogram or a 2D density plot

# Bivariate Analysis

```python
sns.jointplot(x = 'total_bill',y='tip',data=df)
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x = 'total_bill',y='tip',data=df,kind='reg')
```

# Multivariate analysis (we use it when there are more than 2 features)

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```

# DIST PLOTS

- Dist plot helps us to understand the distribution of column features

```python
# if kde is present it will show you density (Kernel Density estimation)
# by default kde is present (is is present is percentage)
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
# if kde is not present it will show you count (Kernel Density estimation)
```
